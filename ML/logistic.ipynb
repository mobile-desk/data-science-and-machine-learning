# Install Modules

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

# Generate Datasets

```python
x, y = make_classification(
    n_samples=100, #how many elements you want
    n_features=1,
    n_classes=2,
    n_clusters_per_class=1,
    flip_y=0.03,
    n_informative=1,
    n_redundant=0,
    n_repeated=0
)
```

```python
print(x)
```

```python
print(y)
```

# Visualize Data

```python
plt.scatter(x, y, c=y, cmap='rainbow')
plt.title('Scatter Plot')
plt.show()

#this should give the s(sigmoid) shape
```

# Split Data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)
```

# Logistic Regression

```python
log_reg = LogisticRegression()
log_reg.fit(x_train, y_train)
print(log_reg)
```

# Make Predictions

```python
y_pred = log_reg.predict(x_test)
```

```python
y_test.shape
```

```python
y_pred.shape
```

# Confusion Matrix

```python
# you have TP(true positive), FP(false positive), FN(false negative), TN(true negative)
# anything false should be low


# ROC(receiver operating characteristics) curve

# Show Coefficient and Intercept
print(f'Coefficient: {log_reg.coef_}')
print(f'Intercept: {log_reg.intercept_}')
```

```python
confusion_matrix(y_test, y_pred)
```
