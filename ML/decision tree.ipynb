# Import

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, classification_report, recall_score
```

# Load Dataset

```python
data = load_breast_cancer()

#convert
dataset = pd.DataFrame(data=data['data'], columns=data['feature_names'])
dataset
```

# Split Data

```python
X = dataset.copy()
y = data['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

# Train Decision Tree

```python
clf = DecisionTreeClassifier(max_depth=4)
# max_depth isn't compulsory, it's just a stopping criteria, amount of branches on the 3
clf = clf.fit(X_train, y_train)
```

## Parameters

```python
clf.get_params()
```

# Make Predictions

```python
# They are two ways to do this
```

## Getting predictions without giving the y value to the model to see what it comes up with

```python
y # 0 is malignant(i.e cancerous) 1 is benign
```

```python
predictions = clf.predict(X_test)
predictions
```

## shows the probability of each class

```python
# show diff btw tree that doesn't have stopping criteria and the one that does
```

```python
clf.predict_proba(X_test)
# if you don't have a stopping criteria you'll just get [0, 1] or [1, 0] throughout
```

# Check Accuracy

## Accuracy Score

```python
accuracy_score(y_test, predictions)
```

## Confusion Matrix

```python
confusion_matrix(y_test, predictions, labels=[0,1])
```

## Precision Score

```python
precision_score(y_test, predictions)
```

## Recall Score

```python
recall_score(y_test, predictions)
```

## Classification Report

```python
print(classification_report(y_test, predictions, target_names=['malignant', 'benign']))
```
